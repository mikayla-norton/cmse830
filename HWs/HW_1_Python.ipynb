‚úçüèº **Problem 6**: Describe in your own words, in a new markdown cell, what a tuple, a list and an array are in Python. Why do we need all three of these? How are they similar? How are they different? What are the use cases for each?

Tuple: immutable collections of data that can contain any type of data (stored within parentheses)
List: mutable, ordered collections of data that can contain multiple data types and be accessed via index (stored in square brackets)
Array: mutable, ordered collections of the same data types, also accessible via index (stored in square brackets) and handle element by element operations.

We need all three because of the types of information contained in each being used for various purposes. For example, tuples typically create permanent sets, lists typically create vector modifiable sets, and arrays have the option for multiple dimensions beyond scalar and vector that are mutable.

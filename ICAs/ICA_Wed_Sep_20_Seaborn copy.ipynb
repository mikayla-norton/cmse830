```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data.csv')
df
```

```python
df.head()
```

```python
# Dropping the unnamed column as it has all Nan values
df.drop(['Unnamed: 32'], axis=1)
```

```python
df.describe()
```

```python
#Plotting the bargraph for the count of people haveing Benign and Malignant of total people.

df_val_cnt= df.diagnosis.value_counts()
df_val_cnt.plot(kind="bar", width=0.1, color=["red", "blue"], legend=1, figsize=(8, 5))
plt.xlabel("(0 = Benign) (1 = Malignant)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(fontsize=12);
plt.yticks(fontsize=12)
plt.legend(["Benign"], fontsize=12)
plt.show()
```

```python
# Generating the heatmap for the dataset with respect to the columns.
plt.figure(figsize =(30,30))
sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True)
plt.show()
```

```python
sns.set_theme(style="white")
corr = df.corr()
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))
sns.heatmap(corr, mask=mask,square=True, linewidths=.5, cbar_kws={"shrink": .6})
```

Observations: In a correlation hat map, the way its interpreted is higher the correlation value, the more correlated the two features are:

- Radius, Area, and Perimeter are all associated (corr>0.9), which is understandable given that area and perimeter are derived using radius data.    
-Texture_mean and texture_worst are highly connected, with corr_value = 0.98 (texture_worst has the highest value of all textures).      
- Compactness_mean,concavity_mean,concave_points_mean are also highy correlated with values in range 0.7 to 0.9.       
- Symmetry_mean and symmetry_worst are correlated too by values 0.7.        
- Fractural_dimension_mean and fractural_dimension_worst are correlated by value 0.77
